<h2 class="section-heading centered">the eeg in epilepsy</h2>

<p class="paragraph">Epileptiform activity on EEG is indicative of cortical hyperexcitability, which carries an increased risk for seizures and the presence of an epileptic network within the brain. There are several types of epileptiform activity, including single discharges (sharps and spikes) and rhythmic and/or periodic activity. It is important to note that the presence of <strong>epileptiform activity is not sufficient for a diagnosis of epilepsy in a patient with no history of seizures</strong>. In a patient with a history of seizures, epileptiform activity in between seizures (interictal activity) is useful in helping to localize seizure onset.<br></p>

<div class="learning-card w-container"><h2 class="heading-12-copy">sharps, spikes &amp; slow waves</h2><p class="paragraph">Any
 epileptiform discharge is a disruption of the usual functioning of the 
brain, and sharps and spikes are perhaps the most classic type. A <strong>sharp </strong>is
 a single epileptiform discharge defined by its duration lasting between
 70-200ms, and by its disruption of the EEG background. A <strong>spike </strong>is
 very similar to a sharp but faster, with a duration from 20-70ms. 
Realistically, the cutoff between them is not very important clinically,
 as one isn't known to be any more "severe" than the other.<br></p><a href="#" class="lightbox-link-8 w-inline-block w-lightbox" aria-label="open lightbox" aria-haspopup="dialog"><img src="Epileptiform%20Activity%20on%20EEG_files/602571bf31ccae53d7ef0c5a_5e78f8a2ce462d41dbb50b47_Left%2520Tem.png" sizes="(max-width: 479px) 96vw, (max-width: 767px) 84vw, (max-width: 991px) 648px, 860px" srcset="Epileptiform%20Activity%20on%20EEG_files/602571bf31ccae53d7ef0c5a_5e78f8a2ce462d41dbb50b47_Left%252_002.png 500w, Epileptiform%20Activity%20on%20EEG_files/602571bf31ccae53d7ef0c5a_5e78f8a2ce462d41dbb50b47_Left%2520Tem.png 1696w" alt="" class="image-17"><script type="application/json" class="w-json">{

<div class="learning-card w-container"><h2 class="heading-12">localizing the discharge</h2><p class="paragraph-3">Generally
 speaking, the location of an epileptiform (or interictal) discharge 
suggests cortical irritability of that region. For example, a discharge 
at T4 suggests right mid-temporal hyperexcitability / epileptogenic 
potential. However, there are a few fine details or caveats to this 
rule.<br><br>First of all, <strong>generalized discharges</strong> 
cannot be localized. Typically found with primary or symptomatic 
generalized epilepsies, generalized discharges begin in such a 
widespread fashion that they effectively involve the entire cortex 
simultaneously. It is possible that deeper structures, such as the 
thalamus, may be involved with generalized discharges.<br></p><img src="Epileptiform%20Activity%20on%20EEG_files/602572185c809ebff6f291d3_5e7d3580322eee2207fc56c6_Untitled_A.png" alt="" class="image-39"><p class="paragraph-3">Note
 that on EEG generalized discharges do not have to be completely the 
same in every single lead; there is often an anterior predominance to 
them, for example, but as long as the morphology of the discharge 
remains throughout all the leads (even if some are less well formed or 
lower amplitude)&nbsp;and the time of onset is the same in all the 
leads, you should consider it a generalized discharge.<br><br>You may also see discharges that appear generalized but are actually<strong> focal with rapid bisynchrony</strong>,
 in which they actually arise from a single location but the networks 
involved propagate the signal too quickly to trace that location, and 
they appear generalized. Sometimes with rapid bisynchrony, with careful 
review you can find a small precedent change before the discharge, such 
as very low amplitude fast activity, to suggest a lateralized onset.<br>‍<br>The
 frontal lobes can be a confusing place for interictal discharges. Deep 
frontal discharges can be missed on scalp EEG entirely, and mesial 
frontal discharges can appear on EEG as if they are from the 
contralateral frontal lobe due to the direction of the discharges’ 
dipole. This isn’t always the case, but something to keep in mind when 
viewing frontal discharges.<br></p><img src="Epileptiform%20Activity%20on%20EEG_files/602572249eab9968d51c50e3_5e7bbf99be9fb2828219d630_mesial%2520f.png" alt="" class="image-37"><p class="paragraph-3"><br>Furthermore,
 while anterior temporal interictals are perhaps the most common focal 
epileptiform discharges, because of the proximity of the F7 and F8 
electrodes to the inferior frontal region, what appear to be anterior 
temporal discharges can, at times, actually be from the inferior frontal
 region.<br></p><img src="Epileptiform%20Activity%20on%20EEG_files/6025722d49c810c2f5d73a2e_5e7bbcb8be9fb2008e19b34b_Untitled_A.png" alt="" class="image-36"><p class="paragraph-3"><br>The
 central region is less complicated: discharges here are usually normal 
in the asleep state (ex. vertex waves from the central region) but 
basically never normal in the awake state. The one caveat is that very 
mesial discharges from the frontal or parietal lobes can sometimes be 
seen on the midline EEG electrodes, so if you see midline discharges in 
the asleep state that are not as symmetric as vertex waves, or with an 
atypical formation or field for vertex waves, you should be suspicious.<br><br>In
 the occipital region, remember that the O1 and O2 electrodes, similar 
to Fp1 and Fp2, suffer from the end of chain issue in bipolar montages. 
That is to say, because there is no electrode behind them to compare 
their voltage to, there is no phase reversal to stand out on EEG. So, if
 you see a positive discharge in the occipital regions that is not 
consistent with a lambda wave or POST, check a circumferential or 
referential montage to clarify if you’re seeing a true interictal 
discharge.<br></p><img src="Epileptiform%20Activity%20on%20EEG_files/6025723f06e7d5c70e351470_5e7bba8973985be7bbc9e3fe_end%2520of%252.png" alt="" class="image-35"></div>

<div class="learning-card w-container"><h2 class="heading-12">Periodic patterns</h2><p class="paragraph-3">Generally
 speaking, periodicity on EEG is almost always suspicious. The 
background should be a little chaotic looking, because all the neurons 
are going about their own business within their respective networks. 
When the background becomes too organized--including via 
periodicity--this raises suspicion for abnormal synchronous firing of 
neurons.<br><br>Periodicity, as defined by the ACNS&nbsp;criteria, requires the relevant discharges to occur for at least <strong>6 cycles</strong>.
 So, if the discharges are periodic at 1 per second, they must go on for
 6 seconds; if they are periodic at 2 discharges per second, they must 
go on for at least 3 seconds; if they are periodic at 3 discharges per 
second, they must go on for at least 2 seconds, and so on.<br></p><img src="Epileptiform%20Activity%20on%20EEG_files/602572743950b6620af84af7_5e7bea6247e8ee7c4b009ef0_Periodicit.png" alt="" class="image-38"><p class="paragraph-3">If the interval between each discharge in a periodic set of discharges varies by 25-50%, it is termed <strong>quasi-periodic</strong>.
 If the variance between each one is more than 50%, it isn't actually 
periodic at all. Full details of the ACNS approved documentation of 
periodic and rhythmic patterns on EEG are available <a href="https://www.acns.org/pdf/guidelines/Guideline-14-pocket-version.pdf" target="_blank">here</a>.<br><br>Below
 is an example of lateralized periodic discharges (LPDs), also termed 
periodic lateral epileptiform discharges (PLEDs). This example may be 
borderline between periodic and quasi periodic, as several of the 
discharges have longer intervals between them bordering on a 
25%&nbsp;difference discharge to discharge.<br></p><a href="#" class="lightbox-link-9 w-inline-block w-lightbox" aria-label="open lightbox" aria-haspopup="dialog"><img src="Epileptiform%20Activity%20on%20EEG_files/60257285e97f310d6e42c8bb_5e7b77b9b6ce77270693f2f6_PLEDs_1.png" sizes="(max-width: 479px) 96vw, (max-width: 767px) 84vw, (max-width: 991px) 648px, 860px" srcset="Epileptiform%20Activity%20on%20EEG_files/60257285e97f310d6e42c8bb_5e7b77b9b6ce77270693f2f6_PLEDs_1-p-.png 500w, Epileptiform%20Activity%20on%20EEG_files/60257285e97f310d6e42c8bb_5e7b77b9b6ce77270693f2f6_PLEDs__002.png 800w, Epileptiform%20Activity%20on%20EEG_files/60257285e97f310d6e42c8bb_5e7b77b9b6ce77270693f2f6_PLEDs_1.png 1600w" alt="" class="image-18"><script type="application/json" class="w-json">{

<div class="learning-card w-container"><h2 class="heading-12">rhythmic patterns</h2><p class="paragraph-3">Recall
 that some rhythmic activity, such as frontal intermittent rhythmic 
delta activity (FIRDA) or generalized intermittent rhythmic delta 
activity (GIRDA), is not necessarily epileptogenic. When you see a <strong>consistently lateralized rhythmic pattern</strong> (such as theta or delta), however, this should raise suspicion for focal cortical hyperexcitability.<br><br>Rhythmic
 patterns carry similar requirements to periodic patterns, needing at 
least 6 cycles to be considered a rhythmic pattern. While periodic 
patterns, however, have discrete intervals between each discharge, 
rhythmic patterns do not--one waveform flows directly into the next.<br>‍<br><strong>Temporal intermittent rhythmic delta activity</strong>
 (TIRDA), in particular, is an epileptogenic pattern often associated 
with an underlying focal temporal lesion such as a tumor or cortical 
dysplasia.<br></p><a href="#" class="w-inline-block w-lightbox" aria-label="open lightbox" aria-haspopup="dialog"><img src="Epileptiform%20Activity%20on%20EEG_files/5fbbe45b1591fe9d8c4ca3f6_tinym%20left%20TIRDA%20again.png" alt=""><script type="application/json" class="w-json">{

fast activity over the frontal regions is usually muscle artifact.

most commonly seen in patients with generalized epilepsy, Lennox-Gastaut
 Syndrome, and tonic seizures.<br><br>Paroxysmal fast activity can be

<div class="learning-card w-container"><h2 class="heading-12">brief potentially ictal rhythmic discharges (birds)</h2><p class="paragraph-3">If
 you look up in the sky you may see birds, but you can also see them on 
EEG&nbsp;tracings. Brief potentially ictal rhythmic discharges, or 
B(I)RDs, are ictal appearing rhythms that evolve but <em>do not meet the minimum 10 second criteria</em>
 to be considered a seizure. They can be seen in a broad range of 
patients, but most commonly in critically ill patients (especially those
 with newfound tumors and strokes), neonates, and sometimes those with 
NMDA&nbsp;receptor encephalitis.&nbsp;<br><br>While B(I)RDs are not 
technically seizures, they are associated with a similarly poor outcome 
and thus merit treatment; in fact, over&nbsp;90% of patients with 
B(I)RDs on EEG&nbsp;will go on to have seizures. In the example below, 
we see an ~8 second run of evolving left temporal spike wave activity 
consistent with B(I)RDs, and note the field into the left parasagittal

<div class="learning-card w-container"><h2 class="heading-12">photoparoxysmal Response</h2><p class="paragraph-3">The <strong>photoparoxysmal response</strong>,
 while perhaps one of the most well known facets of epilepsy, is in 
actuality quite rare. It describes the emergence of interictal activity 
as a result of photic stimulation. Typically, this interictal activity 
is generalized or focal occipital, arises during but outlasts the photic
 stimulation itself, and is always reproducible at the same frequency of
 light flashes. Note that photic driving does not confer an increased 
risk for a photoparoxysmal response or photosensitive seizures.<br></p></div>
